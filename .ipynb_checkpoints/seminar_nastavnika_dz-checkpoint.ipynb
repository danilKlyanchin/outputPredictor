# SARIMAX

```python
# Проверка стационарности
result = adfuller(df['выход'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
# Если ряд не стационарен, применяем дифференцирование
data_diff = df['выход'].diff().dropna()
data_diff
```

```python
# Подбор параметров p, d, q
# Здесь можно использовать ACF и PACF для выбора p и q

p = 2 # autoregression order
d = 1 # num of diff
q = 1 # order of moving average
```

```python
# Создание и обучение модели
model = SARIMAX(df['выход'], order=(p, d, q))
results = model.fit()
```

```python
# Прогнозирование
num_steps = 30
forecast = results.get_forecast(steps=num_steps)
forecast_index = pd.date_range(start=df['выход'].index[-1], periods=1 + num_steps, freq='M')[1:]
forecast_values = forecast.predicted_mean
```

```python
# Визуализация результатов
plt.figure(figsize=(12, 6))
plt.plot(df['выход'], label='Historical Data')
plt.plot(forecast_index, forecast_values, label='Forecast', color='red')
plt.fill_between(forecast_index,
                 forecast.conf_int()['lower выход'],
                 forecast.conf_int()['upper выход'], color='pink')
plt.legend()
plt.show()
```
